```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()

np.random.seed(0)

path = '../data/filtered_data/'

electrodes = np.load(path + 'electrodes.npy')
signals = np.load(path + 'signals_separated.npy')
targets = np.load(path + 'targets_separated.npy')
```

```python
idx = np.random.randint(signals.shape[0])
idx2 = np.random.randint(signals.shape[1])
no_channels = 64

fig, ax = plt.subplots(no_channels, 1, figsize=(16,60), sharex=True)
fig.suptitle('Signals for ' + targets[idx,idx2], fontsize=20)
for i in range(no_channels):
    ax = plt.subplot(no_channels,1, i+1)
    ax.plot(signals[idx,idx2,i,:])
    ax.set_ylabel(electrodes[i])

fig.tight_layout()
fig.subplots_adjust(top=0.97)
plt.show()
```

```python
def one_hot(array):
    unique, inverse = np.unique(array, return_inverse=True)
    onehot = np.eye(unique.shape[0])[inverse]
    return onehot, unique

one_hot(targets)
```
