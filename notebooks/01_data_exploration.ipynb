```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()

np.random.seed(0)

path = '../data/filtered_data/'

electrodes = np.load(path + 'electrodes.npy')
signals = np.load(path + 'signals.npy')
targets = np.load(path + 'targets.npy')
```

```python
idx = np.random.randint(signals.shape[0])
no_channels = 64

fig, ax = plt.subplots(no_channels, 1, figsize=(16,60), sharex=True)
fig.suptitle('Signals for ' + targets[idx], fontsize=20)
for i in range(no_channels):
    ax = plt.subplot(no_channels,1, i+1)
    ax.plot(signals[idx,i,:])
    ax.set_ylabel(electrodes[i])

fig.tight_layout()
fig.subplots_adjust(top=0.97)
plt.show()
```
